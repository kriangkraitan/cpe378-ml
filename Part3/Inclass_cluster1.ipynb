# df1

### Silhouette coefficients and plot

```python
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans
for k in range(2,10):
    labels = KMeans(n_clusters=k).fit(df1).predict(df1)
    print("K = {}, Silhouette score = {:1.4f}".format(k,silhouette_score(df1, labels)))
```

```python
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans
for k in range(2,3):
    labels = KMeans(n_clusters=k).fit(df1).predict(df5)
    print("K = {}, Silhouette score = {:1.4f}".format(k,silhouette_score(df5, labels)))
```

### Fit K-means model

```python
from sklearn.cluster import KMeans

# Number of clusters
K = 4

# Fit the input data
kmeans = KMeans(n_clusters=K).fit(df1)

# Get the cluster labels
labels = kmeans.predict(df1) # or use kmeans.labels

# Get sum of squares distance of all points
sse = kmeans.inertia_

# Centroid values
centroids = kmeans.cluster_centers_

np.set_printoptions(precision=3)
print("Sum of squared errors : {:.3f}".format(sse))
print("\nCentroids : \n", centroids)
```

### Visualize the clustering 1

```python
import matplotlib.pyplot as plt

plt.scatter(df1['X1'], df1['X2'], c=labels, s=5, cmap='Accent')

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5);
```
